```python
ingredients = [['Apple', 'Apples'], ['Artichoke'], ['Winter Melon'], 
         ['Asparagus'], ['Avocado'], ['Bacon'], ['Bamboo Shoot', 'Bamboo Shoots'], 
         ['Banana', 'Bananas'], ['Bean', 'Beans', 'Red Bean', 'Red Beans', 'Black Bean', 'Black Beans', 'Long Bean', 'Long Beans'],
         ['Beef'], ['Beetroot'], ['Bitter Gourd'], ['Black Lentil', 'Black Lentils'], 
         ['Calabash'], ['Bread'], ['Eggplant'], ['Broccoli'], ['Buffalo Meat'], 
         ['Butter'], ['Cabbage'], ['Capsicum'], ['Carrot', 'Carrots'], 
         ['Cassava'], ['Cauliflower'], ['Chayote'], ['Cheese'], 
         ['Chicken', 'Chicken Gizzard'], ['Chili Pepper', 'Chili Peppers'], ['Bell Pepper', 'Bell Peppers'],
         ['Chili Powder'], ['Chow mein'], ['Cinnamon'], ['Coriander'], 
         ['Corn'], ['Crab Meat'], ['Cucumber'], ['Egg', 'Eggs'], ['Fish'], ['Garden cress'], 
         ['Garlic'], ['Ginger'], ['Brinjal'], ['Green Lentil'], ['Mint'], 
         ['Pea', 'Peas', 'Garden Pea', 'Garden Peas', 'Chickpea', 'Chickpeas', 'Soybean', 'Soybeans', 'Pigeon pea', 'Pigeon Peas'],
         ['Edamame'], ['Ham', 'Hams'], ['Jackfruit'], ['Lemon', 'Lemons'], 
         ['Lime', 'Limes'], ['Masyaura'], ['Milk'], ['Minced Meat'], ['Mushroom', 'Mushrooms'], ['Mutton', 'Muttons'], 
         ['Okra'], ['Olive Oil'], ['Onion', 'Onions'], ['Green Onion', 'Green Onions'], ['Orange', 'Oranges'], 
         ['Spinach'], ['Paneer'], ['Papaya'], ['Pear', 'Pears'], ['Pointed Gourd'], ['Pork'], 
         ['Pumpkin', 'Pumpkins', 'Pumpkin shoot', 'Pumpkin shoots'], ['Radish'], ['Mustard Green'], 
         ['Red Lentil', 'Red Lentils'], ['Yellow Lentil', 'Yellow Lentils'], 
         ['Rice'], ['Sausage', 'Sausages'], ['Snake Gourd'], ['Strawberry', 'Strawberries'], 
         ['Potato', 'Potatoes', 'Sweet Potato', 'Sweet Potatoes'], ['Taro', 'Taro Root'], ['Tofu'], ['Tomato', 'Tomatoes'], ['Tamarillo'], 
         ['Turnip'], ['Walnut', 'Walnuts'], ['Watermelon'], ['Wheat', 'Wheats'], ['kimchi'], ['noodle', 'noodles'], ['seaweed', 'seaweeds'],
         ['tumeric']]

products = [
            ['casserole'], ['soup'], ['dessert'], ['salad'], ['noodle'], ['curry'], 
            ['smoothie'], ['drink'], ['stew'], ['sandwich'], ['pasta'], 
            ['chutney'], ['frittata'], ['omelet', 'omelette'], ['cookies'],
            ['meatloaf'], ['dip'], ['pie'], ['brownies'], ['lasagna'],
            ['side dish'], ['breakfast'], ['lunch'], ['potato hash', 'sweet potato hash'], ['hash brown', 'hash browns'],
            ['French Fries'], ['Spaghetti'], ['sauce'], ['barbeque'], 
            ['boil', 'boiled', 'boiling'], ['steam', 'steamed', 'steaming'], ['bake', 'baked', 'baking'],
            ['grill', 'grilled', 'grilling'], ['roast', 'roasted', 'roasting'], ['braise', 'braised', 'braising'],
            ['seared', 'sear', 'searing', 'pan-seared', 'pan sear', 'pan searing'], 
            ['fry', 'fried', 'frying', 'stir-fry', 'stir fry', 'stir fried', 'stir-fried', 'stir frying', 'stir-frying']
        ]

label1 = 'ingredient'
label2 = 'product'
data = []
```

```python
with open(r'C:\Users\user\Desktop\Coding_shit\Python\DL\recipes\NER\Training_data.txt', 'r+') as f:
    
    text = f.readlines()
    random.shuffle(text)
    f.seek(0)
    f.writelines(text)
    f.truncate()

    for line in text:
        temp = {}
        temp['text'] = line
        temp['entities'] = []

        for ingredient_list in ingredients:
            longest = 0
            longest_start = 0

            for ingredient in ingredient_list:
                start = line.lower().find(ingredient.lower())

                if start < 0:
                    continue

                n = len(ingredient)    
                if n >= longest: #greedy
                    longest = n
                    longest_start = start

            if longest != 0:
                temp['entities'].append((longest_start, longest_start+longest, label1))
               
        for product_list in products:
            longest = 0
            longest_start = 0

            for i, product in enumerate(product_list):
                start = line.lower().find(product.lower())

                if start < 0:
                    continue

                n = len(product)    
                if n >= longest: #greedy
                    longest = n
                    longest_start = start

            if longest != 0:
                temp['entities'].append((longest_start, longest_start+longest, label2))

        data.append(temp)
```

```python
train = data[:180]
val = data[180: 204]
test = data[204:]
```

```python
nlp = spacy.blank('en')
train_doc_bin = DocBin()
val_doc_bin = DocBin()
```

```python
for d, doc_bin in [(train, train_doc_bin), (val, val_doc_bin)]:
    for t in d:
        text = t['text']
        labels = t['entities']
        doc = nlp.make_doc(text)
        entss = []
        for start, end, label in labels:
            span = doc.char_span(start, end, label= label)
            if span is None:
                print('Skipping entity')
            else:
                entss.append(span)
        entss = filter_spans(entss)
        doc.ents = entss
        doc_bin.add(doc= doc)

train_doc_bin.to_disk('train.spacy')
val_doc_bin.to_disk('val.spacy')
```
